```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
plt.style.use('seaborn')
```

```python
# Parameter setting

N, n_feature  = 100, 3  # the number of DATA, features
lr = 0.05               # Learning RATE

#target weigt & bias (colunm Vector)
t_w, t_b = np.random.uniform(-5,5,(n_feature,1)), np.random.uniform(-3,3, (1,1))

# initial weight & bias (colunm Vector)
w, b = np.random.uniform(-5,5,(n_feature,1)),np.random.uniform(-3,3, (1,1))
```

```python
# Data Generation

x_data = np.random.randn(N,n_feature)      # N x n_feature
y_data = x_data @ t_w + t_b                # (N,n_feature) x (n_feature, 1) + (1,1)
y_data = y_data + np.random.rand(N,1)  # Noise
# Shape check 
print('input  shape            : ', x_data.shape)
print('weight shape            : ', w.shape)
print('bias   shape            : ', b.shape)
print('output shape            : ', y_data.shape)
```

```python
# for Loss, Weight, Bias Tracking
J_track = list()
w_track, b_track = list(), list()

for data_idx, (x,y) in enumerate(zip(x_data, y_data)):
    #print('data & Label shape      : ', x.shape, y.shape)
    w_track.append(w)
    b_track.append(b)
    
    # forward
    x = x.reshape(1,-1)   # convert dim2 to calculate with bias (shape (1,1))
    pred = x @ w + b

    J = (y-pred)**2               # MSE LOSS
    J_track.append(J.squeeze())    #  Squeeze for Visulize
    #print('prediction & Loss shape : ', pred.shape, J.shape) # (1,1) & (1,1)

    # jacobians
    dJ_dpred = -2 *(y-pred)
    dpred_dw = x                   # (1, 3)  ROW Vector (Converted)
    dpred_db = 1

    # backward
    dJ_dw = dJ_dpred * dpred_dw
    dJ_db = dJ_dpred * dpred_db
    #print('update shape            : ', dJ_dw.shape, dJ_db.shape)  # (1, 3)  & (1,1)

    # parameter update
    w = w - lr*dJ_dw.T             # dJ_dw coverted to Column Vector to calculate with weight vector
    b = b - lr*dJ_db

print('\n ## Result')
print('Target Weight \n', t_w)
print('Target Bias \n', t_b)
print('Learned Weight \n', w)
print('Learned Bias \n', b)
```

```python
print(type(w_track), len(w_track))
print(type(w_track[0]), w_track[0].shape)
print(type(b_track), len(b_track))
print(type(b_track[0]), b_track[0].shape)
print('='*70)

w_track = np.hstack(w_track)       # Stack arrays in sequence horizontally (column wise)
b_track = np.concatenate(b_track)   # Join a sequence of arrays along an existing axis.

print(type(w_track), w_track[0].shape,w_track[1].shape,w_track[2].shape)
print(type(b_track), b_track.shape)
```

```python
#Visulize

fig, axes = plt.subplots(2,1,figsize=(20,10))
# LOSS
axes[0].plot(J_track)
axes[0].set_ylabel('MSE LOSS', fontsize=20)
axes[0].tick_params(labelsize=20)

# Weight
cmap = plt.get_cmap('rainbow', lut = n_feature)
for w_idx, (t_w,w) in enumerate(zip(t_w, w_track)):
    axes[1].axhline(y=t_w, color=cmap(w_idx),linestyle=':')
    axes[1].plot(w, color=cmap(w_idx))

# Bias
axes[1].axhline(y=t_b, color='black',linestyle=':')
axes[1].plot(b_track, color='black')
axes[1].tick_params(labelsize=20)
axes[1].set_ylabel('Weight & Bias Tracking', fontsize=20)
```
