```python
import torch
import torch.nn as nn
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

# Odds, Logit ( 확률을 나타내기 위한 또다른 표현)

## Odds : o = p / ( 1 - p )
- P가 0.5이상이면 o는 1이상
- 비대칭적
- ( 0 ~ + 무한대)

## Logit  = log(Odds)  =  log ( p / ( 1 - p ))    =>     Sigmoid와 역함수 관계  
### Probability -> Logit 변환
- P가 0.5이상이면 logit는 0 이상
- 대칭적 
- ( - 무한대 ~ + 무한대)



```python
p_np = np.linspace(0.01, 0.99, 100)

odds_np = p_np / (1-p_np)
logit_np = np.log(odds_np)

fig, axes = plt.subplots(2,1, figsize=(10,5), sharex=True)
axes[0].plot(p_np, odds_np)
axes[1].plot(p_np, logit_np)
xticks = np.arange(0, 1.1, 0.1)
axes[0].tick_params(labelsize=15)
axes[0].set_xticks(xticks)
axes[0].set_ylabel('Odds', fontsize='20')
axes[1].tick_params(labelsize=15)
axes[1].set_xticks(xticks)
axes[1].set_ylabel('Logit', fontsize='20')
axes[1].set_xlabel('Probability', fontsize='20')
```

## Sigmoid : P = 1 / ( 1 + exp ( -logit ) )

### Logit과 역함수 관계 : Logit -> Probability 변환



```python
x = torch.linspace(-30, 30, 100)
sigmoid = nn.Sigmoid()(x)
                       
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x.numpy(), sigmoid.numpy())
```

# Logistic Regression

```python
x = torch.randn(100,1)  # Normal Distribution

dense = nn.Linear(1,1)
activation = nn.Sigmoid()

y = activation(dense(x))
print(y.shape)
fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(x.detach().numpy().flatten(), y.detach().numpy().flatten())
```

```python
x = torch.randn(100,5)     # Normal Distribution

dense = nn.Linear(5,1)     # 1 ouput feature for Logistic Regresstion
activation = nn.Sigmoid()

y = activation(dense(x))
print(x.shape)
print(y.shape)
```

# Modeling

```python
model = nn.Sequential(
          nn.Linear(5,10),
          nn.ReLU(),
          nn.Linear(10,5),
          nn.ReLU(),
          nn.Linear(5,1),
          nn.Sigmoid()
        )

y = model(x)
print(x.shape)
print(y.shape)
```
