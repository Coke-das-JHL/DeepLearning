## Sigmoid Function
 입력을 (0,1) 사이의 값으로 normalize해준다.  
<div>

</div>


```python
import numpy as np
import matplotlib.pyplot as plt
low, high = -10, 10
n_data = 100                          # number of Sampling Points
interval = (high - low)/(n_data - 1)
linspace = [ low+i*interval for i in range(n_data)] 
linspace = np.array(linspace)

# sigmoid
sigmoid = (1/(1 + np.exp(-linspace)))
fig, ax = plt.subplots(figsize=(6, 2))
ax.plot(linspace, sigmoid)
ax.grid()
```

## Tanh Activation Function
 sigmoid fuction을 보완하고자 나온 함수,  
 입력신호를 (−1,1) 사이의 값으로 normalization 해준다. 
<div>

</div>

```python
# set the linear space
low, high = -10, 10
n_data = 100
interval = (high - low)/(n_data - 1)
linspace = [low + data_idx*interval for data_idx in range(n_data)]
linspace = np.array(linspace)

# tanh
temp1= np.exp(linspace) - np.exp(-linspace)
temp2 = np.exp(linspace) + np.exp(-linspace)
tanh= temp1/temp2

fig, ax = plt.subplots(figsize=(6, 2))
ax.plot(linspace, tanh)
ax.grid()
```

## ReLU Graph
현재 가장 많이 사용되는 Activation Function
<div>

</div>

```python
low, high = -10, 10
n_data = 100
interval = (high - low)/(n_data - 1)
linspace = [low + data_idx*interval for data_idx in range(n_data)]
Relu = np.array(linspace)

# relu
Relu[Relu<0] =0 

fig, ax = plt.subplots(figsize=(6, 2))
ax.plot(linspace, Relu)
ax.grid()
```

## Information Graph

+ 정보 => 놀람의 크기(Surprising degree) 
    - 발생 확률이 높다 => 가치가 낮다. /  발생 확률이 낮다 => 가치가 높다.  

<div>

</div>

```python
# set the linear space
low, high = 0.001, 1
n_data = 100
interval = (high - low)/(n_data - 1)
linspace = [low + data_idx*interval for data_idx in range(n_data)]
linspace = np.array(linspace)

# information function
information = -np.log(linspace) 

fig, ax = plt.subplots(figsize=(6, 2))
ax.plot(linspace, information)
ax.tick_params(labelsize=10)
ax.grid()
ax.set_xlabel("Probability", fontsize=10)
ax.set_ylabel("Information", fontsize=10)
```

## Entropy Fuction

+ 확률변수의 불확실성(Uncertainty)을 측정하는 도구 중의 하나
+ 정보량의 기대값으로 정의
+ 엔트로피는 정보의 불확실성이나, 해당 정보를 접했을 때 느끼는 평균적인 파급력(놀람)의 크기 정도로 해석
    - 평균적으로 100% 에 가깝거나 0%에 가까운 경우 불확실성이 적다 == 놀람의 크기가 작다
    - 50% 부근이 가장 불확실성이 크다.
    
    
+ Deep Learning에서 Cross Entropy Function과 연결시켜 Cost Function으로 사용
    - 얻을 수 있는 정보량의 기대값  => 최대화의 대상
    - Cost Function은 최소화의 대상 => 음의 부호를 붙여 최소화의 대상으로 바꿈

<div>

</div>

```python
# set the linear space
low, high = 0.001, 0.999
n_data = 100
interval = (high - low)/(n_data - 1)
linspace = [low + data_idx*interval for data_idx in range(n_data)]
linspace = np.array(linspace)

# binary entropy function
binary_entropy =  -(linspace*np.log(linspace) + (1-linspace)*np.log(1-linspace))

fig, ax = plt.subplots(figsize=(6, 2))
ax.plot(linspace, binary_entropy)
ax.set_xlabel("Probability", fontsize=10)
ax.set_ylabel("Binary Entropy", fontsize=10)
ax.tick_params(labelsize=10)
ax.grid()  
```
