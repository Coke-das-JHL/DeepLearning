```python
import numpy as np
import matplotlib.pyplot as plt

# 1D Zero-padding

# random data
data = np.random.randint(-10,10,(10,))
data_len = len(data)
print("Input data: ", data)
print("Length of data: ", data_len, '\n')\

# zero-padding
pad = 2
padded_data = np.append(np.array([0]),data)
padded_data = np.append(padded_data,np.array([0]))
print("Input data: ", padded_data)
print("Length of data: ", len(padded_data), '\n')
```

```python
# 1D Average Pooling with Zero-padding

# Window Size
win_size = 3

# random data
out_len = len(padded_data) - (win_size - 1)

# average pooling
output = np.zeros((out_len,))
for idx in range(out_len):
    output[idx] = np.mean(padded_data[idx:idx+win_size])

print("Pooled data: ", np.round(output,2))
print("Length of pooled data: ", len(output), " => Same Length with Original Data")
```

```python
# 1D Correlation with Zero-padding

corr_filter = [-1, 1, 1]

# filter size and output length
filter_size = len(corr_filter)

output = np.zeros((out_len,))
for idx in range(out_len):
    # extract the window
    window = padded_data[idx:idx+filter_size]
    output[idx] = np.sum(window*corr_filter)

print("Correlation :  ", output)
print("Output Length: ", len(output), " => Same Length with Original Data")

fig, ax = plt.subplots(figsize=(10, 5))
# 0 for adjusting locations
output = np.append(np.zeros((1,)),output)
ax.plot(data, label='data')
ax.plot(output, label='correlation value')
ax.legend(fontsize=10)
ax.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 2D Zero-padding

# random data
data = np.random.randint(-10,10,(10,10))
data_len = len(data)
print("Input data \n", data)
print("Shape of data: ", data.shape, '\n')

# zero-padding with numpy
padded_data = np.pad(data, (1, 1), 'constant',constant_values=(0, 0))
print("Padded   data:\n", padded_data)
print("Shape of data: ", padded_data.shape, '\n')
```

```python
# zero-padding 
pad = 1
H,W = data.shape
padded_data2 = np.append(       np.zeros((1,W)),                  data, axis=0)
padded_data2 = np.append(          padded_data2,       np.zeros((1,W)), axis=0)
padded_data2 = np.append(          padded_data2, np.zeros((H+pad*2,1)), axis=1)
padded_data2 = np.append( np.zeros((H+pad*2,1)),          padded_data2, axis=1)                     
print("Padded   data:\n", padded_data2)
print("Shape of data: ", padded_data2.shape, '\n')
```

```python
# 2D Max Pooling with Zero-padding

# Window Size
H,W = padded_data.shape
window_size = 3 # 3x3

# Output
output_H, output_W = H - (window_size - 1), W - (window_size - 1)

output = np.zeros((output_H,output_W))
for h in range(output_H):
    for w in range(output_W):
        # extract the window
        window_data = data[h:h+window_size,w:w+window_size]
        output[h,w] = np.max(window_data)
print('=====')

print(np.round(output,3))
print('Output shape: ', output.shape)
```

```python
# 2D Correlation with Zero-padding

# generate the random filter
corr_filter = np.random.randint(0,5,(filter_size,filter_size))
print(corr_filter)
print('=====')

# Output
H,W = padded_data.shape
output_H, output_W = H - (filter_size - 1), W - (filter_size - 1)


output = np.zeros((output_H,output_W))
for h in range(output_H):
    for w in range(output_W):
        # extract the window
        filtered_data = padded_data[h:h+filter_size,w:w+filter_size]*corr_filter
        output[h,w] = np.sum(filtered_data)
print('=====')

print(output)
```
